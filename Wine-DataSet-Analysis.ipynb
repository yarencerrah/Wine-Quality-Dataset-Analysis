```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame = pd.read_csv("WineQT.csv")
```

```python
# checking counts of null value
dataFrame.isnull().sum()
```

```python
# quality values
sns.histplot(x = "quality", data = dataFrame)
```

```python
# duplicate row check
dataFrame[dataFrame.duplicated()]
```

```python
avr_quality = dataFrame.groupby("quality").mean()
avr_quality
```

```python
plt.figure(figsize = (12,7))
sns.heatmap(dataFrame.corr(), annot = True, cmap = "coolwarm")
```

```python
# some of elements effect on quality
plt.figure(figsize = (20,8))
selected_columns = ["volatile acidity","citric acid","chlorides",
                     "total sulfur dioxide","density","sulphates","alcohol", "quality"]
selected_columns2 = ["volatile acidity","sulphates","alcohol", "quality"]
df_selected = dataFrame[selected_columns2]
df_melted = pd.melt(df_selected, id_vars = "quality", var_name = "Feature", 
                    value_name = "Value")
sns.lineplot(x = "Value", y = "quality", hue = "Feature", data = df_melted) 

plt.title("Quality vs Feature Values")
```
